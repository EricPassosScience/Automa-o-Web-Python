# Problema de Negócio e Observações

PROBLEMA - 

Uma importadora trabalha com diversos produtos, produtos esses vinculados a cotação do Dolar, Euro e Ouro. 

Sendo assim, considerando que o preço de tais moedas influencia no faturamento e no preço de compra e venda dos produtos, para melhor visulaizar e agilizar os processos dentro da empresa, a área de dados foi questionada sobre a possibilidade de automatizar a tarefa de atulizar a base de dados com os referidos valores, especialmente a reverberação no preço final dos produtos. 

OBSERVAÇÕES - 

Para resolver esse problema, buscas serão por informações serão feitas na Web, na sequência o banco de dados será atualizado com tais informaçoes e exportado para um arquivo excel, tudo de maneira automatizada.

```python
#2 - importando o selenium e o pandas

from selenium import webdriver #controla o navegador
from selenium.webdriver.common.by import By #selecionar os itens no navegador
from selenium.webdriver.common.keys import Keys #controla o teclado

import pandas as pd
```

```python
#3 - Criando/abrindo o navegador

navegador = webdriver.Chrome()

#3-A: pesquisando e colhendo a cotação do Dolar no google  

navegador.get("https://www.google.com.br/")
navegador.find_element( 
    By.XPATH, '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys("cotação dolar", Keys.ENTER)
cotacao_dolar = navegador.find_element( 
    By.XPATH, '//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute('data-value')
print(cotacao_dolar)

#3-B: pesquisando e colhendo no goggle a cotação do Euro 

navegador.get("https://www.google.com.br/")
navegador.find_element( 
    By.XPATH, '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys("cotação euro", Keys.ENTER)
cotacao_euro = navegador.find_element( 
    By.XPATH, '//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute('data-value')
print(cotacao_euro)

#3-C: Pesquisando e colhendo a cotação do ouro no site: https://www.melhorcambio.com/ouro-hoje

navegador.get("https://www.melhorcambio.com/ouro-hoje")
cotacao_ouro = navegador.find_element(By.XPATH, '//*[@id="comercial"]').get_attribute('value')

cotacao_ouro = cotacao_ouro.replace(',', '.')
print(cotacao_ouro)

#3-D: fechando o navegador

navegador.quit() 
```

```python
#4- importando a base de dados

tabela = pd.read_excel("produtos.xlsx")
display(tabela)
```

```python
#5-A: atualizando a cotação na base de dados

#Função loc[linha, coluna]

tabela.loc[tabela["Moeda"] == "Dólar","Cotação"] = float(cotacao_dolar)

tabela.loc[tabela["Moeda"] == "Euro","Cotação"] = float(cotacao_euro)

tabela.loc[tabela["Moeda"] == "Ouro","Cotação"] = float(cotacao_ouro)

display(tabela)
```

As casas decimais seram melhor apresentadas quando o arquivo for exportado para o excel

```python
#5-B: atulizando o Preço de Compra e Preço de Venda

tabela["Preço de Compra"] = tabela["Preço Original"] * tabela["Cotação"]

tabela["Preço de Venda"] = tabela["Preço de Compra"] * tabela ["Margem"]

display(tabela)
```

```python
#6 - Exportar  a tabela para excel

tabela.to_excel("produtoatualizado.xlsx", index=False)
```
